```python
experiments = comet_api.get("matech96/cifar10-11-batch-size-power2")
n_training_examples = 42000
n_epochs = 10
```

```python
list2float = lambda l: float(l[0])
list2int = lambda l: int(l[0])
durretions = [list2float(comet_api.get_experiment_metrics(experiment.key, 'epoch_duration')) for experiment in experiments]
batch_sizes = [list2int(comet_api.get_experiment_parameters(experiment.key, 'batch_size')) for experiment in experiments]
```

```python
durretions_per_batch = [durretion / (ceil(n_training_examples / batch_size) * n_epochs) for durretion, batch_size in zip(durretions, batch_sizes)]
data = pd.DataFrame({'durretions_per_batch': durretions_per_batch, 'batch_sizes': batch_sizes})
```
