```python
from keras.datasets import cifar10
from bokeh.plotting import figure, output_notebook, show
from bokeh.palettes import Category10_10
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
labels_str_arr = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
print('Shapes:')
for name, tensor in zip(('x_train', 'y_train', 'x_test', 'y_test'), (x_train, y_train, x_test, y_test)):
    print(name, tensor.shape)
```

```python
print('Number of examples: training: {:,}, test: {:,}'.format(x_train.shape[0], x_test.shape[0]))
```

```python
import numpy as np
_, train_counts = np.unique(y_train, return_counts=True)
_, test_counts = np.unique(y_test, return_counts=True)
data = {'label': labels_str_arr, 'train': train_counts, 'test': test_counts}
```

```python
width = 800
height = 200

output_notebook()
p = figure(title='category distribution (train)', x_range=labels_str_arr, plot_width=width, plot_height=height)
p.vbar_stack(('train', 'test'), x='label', source=data, width=0.9, color=Category10_10[:2])
show(p)
```
